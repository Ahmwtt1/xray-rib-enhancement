```python
def display_comparison(original_img, processed_img, title):
    """Orijinal ve işlenmiş görüntüyü yan yana karşılaştırmalı gösterir."""
    plt.figure(figsize=(12, 6))

    # Orijinal Görüntü
    plt.subplot(1, 2, 1)
    plt.title("ÖNCESİ: Orijinal Görüntü")
    plt.imshow(original_img, cmap='gray')
    plt.axis('off')

    # İşlenmiş Görüntü (YÖNTEM ÇIKTISI)
    plt.subplot(1, 2, 2)
    plt.title(f"SONRASI: {title}")
    plt.imshow(processed_img, cmap='gray')
    plt.axis('off')

    plt.show()

def apply_clahe(img, file_name):
    """1. Histogram Eşitleme (CLAHE) uygular ve gösterir."""
    clahe = cv2.createCLAHE(clipLimit=2.0, tileGridSize=(8, 8))
    img_clahe = clahe.apply(img)
    display_comparison(img, img_clahe, f"1. CLAHE (Kontrast Artırma) - {file_name}")
    return img_clahe

def apply_sobel(img, file_name):
    """2. Sobel Filtresi uygular ve gösterir."""
    sobelx = cv2.Sobel(img, cv2.CV_64F, 1, 0, ksize=3)
    sobely = cv2.Sobel(img, cv2.CV_64F, 0, 1, ksize=3)
    img_sobel = cv2.magnitude(sobelx, sobely)
    img_sobel = cv2.convertScaleAbs(img_sobel)
    display_comparison(img, img_sobel, f"2. Sobel Filtresi (Kenar Tespiti) - {file_name}")
    return img_sobel

def apply_hpf(img, file_name):
    """3. Yüksek Geçiren Filtre (HPF) uygular ve gösterir."""
    kernel_size_hp = 25
    img_low_pass = cv2.blur(img, (kernel_size_hp, kernel_size_hp))
    # HPF = Orijinal - Low Pass
    img_high_pass = cv2.addWeighted(img, 1.0, img_low_pass, -1.0, 128)
    display_comparison(img, img_high_pass, f"3. Yüksek Geçiren Filtre (HPF) - {file_name}")
    return img_high_pass

def apply_bilateral(img, file_name):
    """4. Bilateral Filtreleme uygular ve gösterir."""
    img_bilateral = cv2.bilateralFilter(img, 9, 75, 75)
    display_comparison(img, img_bilateral, f"4. Bilateral Filtreleme - {file_name}")
    return img_bilateral
```
